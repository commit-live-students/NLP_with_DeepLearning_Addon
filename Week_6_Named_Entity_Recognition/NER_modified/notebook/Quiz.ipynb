2. What is the regex for finding a single digit in 'He returned home after 7 years' ?

    a) .*
    
    b) .+
    
    c)[a-z]*
    
    d) [0-9]
    
__Ans__: d

__Explanation__: As discussed, the bracket defines a restricted scope for search. Although a) will also select the digit, but also all the other characters in the sentence.

3. What is the regex to search for lamb in 'Mary had a little lamb' ?

    a)lamb
    
    b)l.*b
    
    c)l.+b
    
    d)l.{2}b
    
__Ans__: a) and d). However, d counts for points in the answer.

__Explanation__:
    * a) because its a simple text match. No brainer.
    * d) because it exactly outlines that l, followed by 2 characters(a,m) and then b. b) and c) would both select little lamb. Think about it.
    
4. Select the regex to find everyting before the first comma in a sentence.

    a) ,
    
    b) ,.*
    
    c) ,*
    
    d) .+,
    
__Ans__: d)

__Explanation__: .+, means that select one or more characters in the sentence up till ,(comma) which is required statement of the problem.

5. Select one word after and one word before all the occurances of the word "and" in a sentence.

__Ans__: (\S+\s)?and\s(\S+)

__Explanation__:

    * Grouping - By placing a regex inside round brackets, we essentially group them together. Any multiplier added is essentially meant for the whole group.
    * \S selects all the non-white space character
    * \s selects all whitespace character
    * The first group, (\S+\s)? says select all non-whitespace and white space character untill you get the word "and", then select the next white-space character(\s, the space after and), and then select one or more non whitespace character (the next word after "and").
